```python
import zipfile
import io
data=zipfile.ZipFile(io.BytesIO(uploaded['pandas-tigres.zip']),'r')
data.extractall()
```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
ruta_tigre="/content/tigres"
Tiger=[]
img_size=150
for img in os.listdir(ruta_tigre):
  img=cv2.imread(os.path.join(ruta_tigre,img))
  #img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize=cv2.resize(img,(img_size,img_size))
  Tiger.append(img_resize)
Tiger=np.array(Tiger)
print(Tiger.shape)

```

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
ruta_panda="/content/pandas"
pandita=[]
img_size=150
for img in os.listdir(ruta_panda):
  img=cv2.imread(os.path.join(ruta_panda,img))
  #img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img_resize=cv2.resize(img,(img_size,img_size))
  pandita.append(img_resize)
pandita=np.array(pandita)
print(pandita.shape)
```

```python
plt.figure()
plt.imshow(np.squeeze(Tiger[0]))
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.figure()
plt.imshow(np.squeeze(pandita[0]))
plt.colorbar()
plt.grid(False)
plt.show()
```

# Nuevo

```python
imagenes=np.concatenate([Tiger, pandita])
print(len(imagenes))
Imagenes=np.array(imagenes)
print(Imagenes.shape)
```

```python
etiquitasTigres=np.repeat(0,12)
print(len(etiquitasTigres))
print(etiquitasTigres)
```

```python
etiquitasPandas=np.repeat(1,15)
print(len(etiquitasPandas))
print(etiquitasPandas)
```

```python
class_names=['Tigre','Panda']
```

```python
labels=np.concatenate([etiquitasTigres,etiquitasPandas])
print(len(labels))
print(labels)
Labels=np.array(labels)
print(Labels.shape)
```

```python
plt.figure(figsize=(10,10))
for i in range(20):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(Imagenes[i])
  plt.xlabel(class_names[Labels[i]])
plt.show()
```
